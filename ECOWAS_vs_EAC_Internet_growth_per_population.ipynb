```
#creating a dataframe with two columns. one for countries from EAC(East African Community),ECOWAS(Economic Community of West African States)  and another column for growth of internet users by percentage population
import pandas as pd
import pandas as DataFrame
data = pd.read_excel('africa.xlsx')
df = pd.DataFrame(data, columns=['AFRICA','growth%pop'])

print(df)
```

```
#creating data and labels for the pie chart

from matplotlib import pyplot as plt

my_data =[252.45,369.42,322.50,162.46,97.07,336.00,199.31,79.00,789.12,662.83,189.31,490.96,242.74,179.49,800.00,204.70,215.64,743.93,19.43,199.00,474.00,8910]

my_labels='Benin','BurkinaFaso','Cabo-Verde','Ivory-Coast','Gambia','Ghana','Guinea','Guinea-Bissau','Liberia','Mali','Niger','Nigeria','Senegal','Sierra Leone','Togo','Burundi','Kenya','Rwanda','South Sudan','Tanzania','Uganda','Rest of World'

```

```
#creating 2 dataframes for the two communities to plot them on different barplots

dataeco=[ ['Benin',252.450],['Burkina Faso',369.420],
['Cabo-Verde', 322.500],
['Ivory-Coast',162.460],
['Gambia',97.070],
['Ghana', 336.000],
['Guinea',199.310],
['Guinea-Bissau',79.000],
['Liberia',789.120],
['Mali',662.830],
['Niger' ,189.310],
['Nigeria',490.960],
['Senegal',242.740],
['Sierra Leone',179.490],
['Togo', 8.000]]  

dfeco=pd.DataFrame(dataeco,columns=['country','internet growth'])  #dataframe for ECOWAS

dataeac=[['Burundi',204.70],['Kenya',215.64],['Rwanda',743.93],['South Sudan',19.43],['Tanzania',199.00],['Uganda',474.00]]

dfeac=pd.DataFrame(dataeac,columns=['country','internet growth']) #dataframe for EAC


print(dfeac)
print(dfeco)
```

```
#PLOTTING two different charts; one for the ECOWAS countries and another one for EAC countries

from itertools import cycle, islice


# my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, len(dfeac)))
dfeac.plot.barh(x ='country',y='internet growth',stacked=True, colormap='Paired')
plt.title("Internet Growth in EAC countries")
plt.xlabel("Internet %per population growth")
plt.ylabel("Countries")

# my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, len(dfeco)))
dfeco.plot.barh(x ='country',y='internet growth',stacked=True, colormap='Paired')
plt.title("Internet Growth in ECOWAS countries")
plt.xlabel("Internet %per population growth")
plt.ylabel("Countries")
```

```
# plotting ECOWAS & EAC countries vs the rest of the world
plt.gca().axis("equal")
pie = plt.pie(my_data,startangle=0, autopct='%1.0f%%', pctdistance=0.9, radius=1.2)
labels=df.index.unique()
plt.title('ECOWAS & EAC vs the rest of the world', weight='bold', size=14)
plt.legend(pie[0],labels=my_labels, bbox_to_anchor=(1,0.5), loc="center right", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
plt.subplots_adjust(left=0.0, bottom=0.1, right=0.50)

plt.show()
plt.clf()
plt.close()
```
